# TIPOS DE DATOS EN PYTHON

Los tipos de datos son para trabajar (manipular) con datos en nuestros proyectos.

Por ejemplo: supongamos que queremos sacar el promedio de calificaciones de un grupo de estudiantes de 3er año de secundaria:

```python
estudiante_1 = 10
estudiante_2 = 8
estudiante_3 = 9
estudiante_4 = 7
estudiante_5 = 5
estudiante_6 = 10

promedio = (estudiante_1 + estudiante_2 + estudiante_3 + estudiante_4 + estudiante_5 + estudiante_6) / 6

print(promedio)
```
